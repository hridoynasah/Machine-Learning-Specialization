### 1️⃣ **House Price Prediction Based on Area**

**Objective:** Predict house prices using the area (square footage) as the sole predictor.

**Dataset:** [USA Housing Dataset](https://www.kaggle.com/datasets/vedavyasv/usa-housing)

**Tasks:**

- Load and visualize the data.

- Implement a simple linear regression model.

- Evaluate the model's performance.

### 2️⃣ **Student Scores Prediction Based on Study Hours**

**Objective:** Predict student scores based on the number of hours studied.

**Dataset:** [Student Performance Dataset](https://www.kaggle.com/datasets/whenamancodes/students-performance-in-exams)

**Tasks:**

- Analyze the relationship between study hours and scores.

- Build and train a linear regression model.

- Make predictions and assess accuracy.

### 5️⃣ **Height Prediction Based on Age for Children**
