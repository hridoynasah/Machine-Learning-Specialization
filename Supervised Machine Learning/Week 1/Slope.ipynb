```python
import numpy as np
import matplotlib.pyplot as plt

# Define the slope (m) and the y-intercept (b)
m = 2  # example slope
b = 1  # y-intercept

# Generate x values
x = np.linspace(-10, 10, 100)

# Calculate the corresponding y values using the equation y = mx + b
y = m * x + b

# Plot the straight line
plt.plot(x, y, label=f'Slope = {m}')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Straight Line with Constant Slope')
plt.legend()
plt.grid(True)
plt.show()
```
