### What is **Supervised Learning**?

Supervised learning is a type of machine learning where an algorithm is trained using examples that include both:

1. **Inputs (x)**: Information you provide to the system, like an email or a picture.
2. **Outputs (y)**: The correct answer or label associated with each input, like "spam" or "not spam."

The algorithm learns to find patterns between these inputs and outputs so that when given new inputs, it can make predictions.

---

### How Does It Work?

1. **Training Phase**:

   - You give the algorithm a lot of paired examples of inputs (x) and correct outputs (y).
   - Over time, the algorithm learns the relationship between x and y.

2. **Prediction Phase**:
   - After learning, the algorithm can take a brand-new input it has never seen before and predict the correct output.

### Examples of Supervised Learning

#### 1. **Spam Filter**:

- **Input (x)**: An email.
- **Output (y)**: "Spam" or "Not Spam."
- The algorithm learns from labeled examples of spam and non-spam emails.

#### 2. **Speech Recognition**:

- **Input (x)**: An audio clip.
- **Output (y)**: Text transcript of the speech.

#### 3. **Machine Translation**:

- **Input (x)**: A sentence in one language (e.g., English).
- **Output (y)**: The translated sentence in another language (e.g., Spanish).

#### 4. **Online Advertising**:

- **Input (x)**: Information about a user and an ad.
- **Output (y)**: Whether the user will click on the ad.
- This is very profitable because predicting clicks increases revenue for companies.

#### 5. **Self-Driving Cars**:

- **Input (x)**: Images and sensor data.
- **Output (y)**: Position of other cars or objects to navigate safely.

#### 6. **Manufacturing Quality Check**:

- **Input (x)**: A picture of a product.
- **Output (y)**: Whether the product has defects.

---

### Types of Supervised Learning Problems

#### 1. **Regression**:

- **What it does**: Predicts a number from infinitely many possible values.
- **Example**: Predicting house prices based on size.
  - **Input (x)**: Size of the house in square feet.
  - **Output (y)**: Predicted price in dollars.
- The algorithm might fit a straight line or a curve to predict the price.

#### 2. **Classification**:

- **What it does**: Assigns the input to one of a fixed number of categories.
- **Example**: Spam or not spam.
  - **Input (x)**: An email.
  - **Output (y)**: "Spam" or "Not Spam."
