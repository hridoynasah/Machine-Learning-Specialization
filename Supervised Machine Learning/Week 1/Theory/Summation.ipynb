## **All You Need To Know About Summation**

#### **Summation notation is a compact way of writing the sum of a series of terms. It uses the Greek capital letter sigma (∑) to indicate that you are summing a sequence of numbers defined by a specific formula.**

### **Basic Structure:**

A typical summation expression looks like this:

$$
\sum_{i=a}^{b} f(i)
$$

- **$\sum$**: This is the summation symbol.
- **$i$**: This is the index of summation; it represents each term's position.
- **$a$**: The lower limit; the summation starts with $i = a$.
- **$b$**: The upper limit; the summation continues until $i = b$.
- **$f(i)$**: This is the general term (or formula) that depends on $i$.

### **Example 1: Simple Sum of Natural Numbers**

To sum the first 5 natural numbers, you write:

$$
\sum_{i=1}^{5} i
$$

This means:

$$
1 + 2 + 3 + 4 + 5 = 15
$$

### **Example 2: Sum of Squares**

Suppose you want to sum the squares of the first 4 numbers:

$$
\sum_{i=1}^{4} i^2
$$

This expands to:

$$
1^2 + 2^2 + 3^2 + 4^2 = 1 + 4 + 9 + 16 = 30
$$

---

### **Properties of Summations**

1. **Linearity:**  
   For any constants $c$ and $d$ and functions $f(i)$ and $g(i)$:
   $$
   \sum_{i=a}^{b} [c \cdot f(i) + d \cdot g(i)] = c \sum_{i=a}^{b} f(i) + d \sum_{i=a}^{b} g(i)
   $$
2. **Splitting the Sum:**  
   You can split a summation over a range into two parts. For example:
   $$
   \sum_{i=1}^{b} f(i) = \sum_{i=1}^{k} f(i) + \sum_{i=k+1}^{b} f(i)
   $$
3. **Changing the Index:**  
   Sometimes you may want to change the index of summation. For example, if you let $j = i - 1$, then:
   $$
   \sum_{i=1}^{n} f(i) = \sum_{j=0}^{n-1} f(j+1)
   $$

## **Properties With Examples:**

### **1. Linearity**

**Property:**

$$
\sum_{i=a}^{b} [c \cdot f(i) + d \cdot g(i)] = c \sum_{i=a}^{b} f(i) + d \sum_{i=a}^{b} g(i)
$$

**Example:**  
Suppose $f(i) = i^2$, $g(i) = i$, $c = 2$, $d = -3$, and the summation is over $i = 1$ to $4$. Then:

$$
\sum_{i=1}^{4} [2 \cdot i^2 - 3 \cdot i]
$$

Using linearity, we can split this into two separate summations:

$$
\sum_{i=1}^{4} [2 \cdot i^2 - 3 \cdot i] = 2 \sum_{i=1}^{4} i^2 - 3 \sum_{i=1}^{4} i
$$

Now calculate each term separately:

- $\sum_{i=1}^{4} i^2 = 1^2 + 2^2 + 3^2 + 4^2 = 1 + 4 + 9 + 16 = 30$
- $\sum_{i=1}^{4} i = 1 + 2 + 3 + 4 = 10$

Substitute these results back:

$$
2 \sum_{i=1}^{4} i^2 - 3 \sum_{i=1}^{4} i = 2(30) - 3(10) = 60 - 30 = 30
$$

Thus, the result is:

$$
\boxed{30}
$$

---

### **2. Splitting the Sum**

**Property:**

$$
\sum_{i=1}^{b} f(i) = \sum_{i=1}^{k} f(i) + \sum_{i=k+1}^{b} f(i)
$$

**Example:**  
Suppose $f(i) = i^2$, and we want to compute $\sum_{i=1}^{5} i^2$. Let’s split the summation at $k = 3$:

$$
\sum_{i=1}^{5} i^2 = \sum_{i=1}^{3} i^2 + \sum_{i=4}^{5} i^2
$$

Now calculate each part:

- $\sum_{i=1}^{3} i^2 = 1^2 + 2^2 + 3^2 = 1 + 4 + 9 = 14$
- $\sum_{i=4}^{5} i^2 = 4^2 + 5^2 = 16 + 25 = 41$

Add the results:

$$
\sum_{i=1}^{5} i^2 = 14 + 41 = 55
$$

Thus, the result is:

$$
\boxed{55}
$$

---

### **3. Changing the Index**

**Property:**  
If you let $j = i - 1$, then:

$$
\sum_{i=1}^{n} f(i) = \sum_{j=0}^{n-1} f(j+1)
$$

**Example:**  
Suppose $f(i) = i^3$, and we want to compute $\sum_{i=1}^{4} i^3$. Let’s change the index by setting $j = i - 1$, so $i = j + 1$. When $i = 1$, $j = 0$, and when $i = 4$, $j = 3$. The summation becomes:

$$
\sum_{i=1}^{4} i^3 = \sum_{j=0}^{3} (j+1)^3
$$

Now expand $(j+1)^3$ and compute the summation:

$$
(j+1)^3 = j^3 + 3j^2 + 3j + 1
$$

So:

$$
\sum_{j=0}^{3} (j+1)^3 = \sum_{j=0}^{3} j^3 + 3 \sum_{j=0}^{3} j^2 + 3 \sum_{j=0}^{3} j + \sum_{j=0}^{3} 1
$$

Calculate each term:

- $\sum_{j=0}^{3} j^3 = 0^3 + 1^3 + 2^3 + 3^3 = 0 + 1 + 8 + 27 = 36$
- $\sum_{j=0}^{3} j^2 = 0^2 + 1^2 + 2^2 + 3^2 = 0 + 1 + 4 + 9 = 14$
- $\sum_{j=0}^{3} j = 0 + 1 + 2 + 3 = 6$
- $\sum_{j=0}^{3} 1 = 1 + 1 + 1 + 1 = 4$

Substitute these results back:

$$
\sum_{j=0}^{3} (j+1)^3 = 36 + 3(14) + 3(6) + 4 = 36 + 42 + 18 + 4 = 100
$$

Thus, the result is:

$$
\boxed{100}
$$

### **A Simple Application**

Suppose you have the summation:

$$
S = \sum_{i=1}^{10} (2i + 3)
$$

You can compute each term and then add them up:

- When $i=1$: $2(1) + 3 = 5$
- When $i=2$: $2(2) + 3 = 7$
- $\vdots$
- When $i=10$: $2(10) + 3 = 23$

So, $S = 5 + 7 + 9 + \dots + 23$.

Alternatively, using properties of summation:

$$
S = \sum_{i=1}^{10} (2i + 3) = 2 \sum_{i=1}^{10} i + 3 \sum_{i=1}^{10} 1
$$

We know:

$$
\sum_{i=1}^{10} i = \frac{10(10+1)}{2} = 55 \quad \text{and} \quad \sum_{i=1}^{10} 1 = 10
$$

Thus,

$$
S = 2(55) + 3(10) = 110 + 30 = 140
$$

### **Summary**

- **Summation notation ($\sum$)** compactly represents the sum of a sequence of terms.
- The expression $\sum_{i=a}^{b} f(i)$ means you sum the values of $f(i)$ as $i$ runs from $a$ to $b$.
- Basic properties such as linearity help break down and simplify more complex summations.
