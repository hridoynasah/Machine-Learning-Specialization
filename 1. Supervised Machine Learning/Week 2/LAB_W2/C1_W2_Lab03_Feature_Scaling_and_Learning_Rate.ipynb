>Note:  
**mu and sigma are "stretched" (broadcasted) to shape (99, 4) during the operation  so each feature gets normalized using its own mean and std.**

```python
# calculating mean for each column/features
mu = np.mean(X_train, axis=0) # axis = 0 means calculating for each column
# calculating standard deviation for each column
sigma = np.std(X_train, axis=0) # axis = 0 means calculating for each column
X_norm = (X_train-mu)/sigma

print(f"Shape of X_norm: {X_norm.shape}")
print(f"Shape of mu    : {mu.shape}")
print(f"Shape of sigma : {sigma.shape}")
```

```python
def zscore_normalize_features(X):
    # calculating mean for each column/features
    mu = np.mean(X, axis=0) # axis = 0 means calculating for each column
    # calculating standard deviation for each column
    sigma = np.std(X, axis=0) # axis = 0 means calculating for each column
    X_norm = (X-mu)/sigma
    
    return X_norm, mu, sigma
```

```python
mu           = np.mean(X_train, axis=0)
sigma        = np.std(X_train, axis=0)
X_minus_mean = (X_train - mu)
X_norm       = (X_train - mu) / sigma

fig, ax = plt.subplots(1, 4, figsize=(12,3), sharey=True)

for i in range(X_norm.shape[1]):
    ax[i].scatter(X_norm[:, i], y_train)
    ax[i].set_xlabel(X_features[i])
ax[0].set_ylabel("Price (1000's)")
plt.show()
```
