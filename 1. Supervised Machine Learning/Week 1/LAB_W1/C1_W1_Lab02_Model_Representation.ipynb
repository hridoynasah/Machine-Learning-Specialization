```python
w = 100
b = 100
print(f"w: {w}")
print(f"b: {b}")
```

```python
x = np.empty(3)
print(x)
```

```python
def compute_output_model(x, y, w, b):
    m = x.shape[0]
    f_wb = np.empty(2)
    for i in range(m):
        f_wb[i] = w * x[i] + b
    
    return f_wb

f_wb = compute_output_model(x_train, y_train, w, b)
print(f_wb)
```

```python
def plot_data(x_train, y_train, f_wb):    
    # Our prediction vs Actual Values 

    # Plot our model prediction 
    plt.plot(x_train, f_wb, c ='b', label = 'Our Prediction')

    # Plot the data points 
    plt.scatter(x_train, y_train, marker='X', c ='r', label = 'Actual Values')

    # Set the title 
    plt.title("Housing prices")
    # Set the y-axis label
    plt.ylabel('Price (in 1000s of dollars)')
    # Set the x-axis label
    plt.xlabel('Size (1000 sqft)')
    plt.legend()
    plt.show()
```

```python
plot_data(x_train, y_train, f_wb)
```
