**1. Importing required libraries:**

```python
import pandas as pd
from matplotlib import pyplot as plt 
plt.style.use("ggplot")
```

**2. Ploting scatter plot:**

```python
x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]

plt.scatter(x, y)
plt.show()
```

**3. Adding color, marker, marker size:**

```python
x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]

# 3. Adding color, marker, marker size:
plt.scatter(x, y, s = 100, color = "green", marker = "X")

plt.title("Random values")
plt.xlabel("Values of x-axis")
plt.ylabel("Value of y-axis")
plt.show()
```

**4. Adding marker edgecolor, edge width, transparency(alpha):**

```python
x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]

# 4. Adding marker edgecolor, edge width, transparency(alpha):
plt.scatter(x, y, s = 100, color = "green", marker = "o",
            edgecolor = 'black', linewidth = 2, alpha = 0.95)

plt.title("Random values")
plt.xlabel("Values of x-axis")
plt.ylabel("Value of y-axis")
plt.show()
```

**5. Adding colormap:**

```python
x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]

colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]

# 4. Adding marker edgecolor, edge width, transparency(alpha), colormapping:
plt.scatter(x, y, s = 100, c = colors, cmap='Greens', marker = "o",
            edgecolor = 'black', linewidth = 2, alpha = 0.95)

cbar = plt.colorbar() 
cbar.set_label("Satisfaction")
plt.title("Random values")
plt.xlabel("Values of x-axis")
plt.ylabel("Value of y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt

# 3 points
x = [1, 2, 3]
y = [4, 5, 6]

# Color values for each point
colors = [10, 5, 1]  # these will map to different shades using the colormap

# Scatter plot with color, edge color, size, alpha (transparency)
plt.scatter(x, y, s=300, c=colors, cmap='Greens', marker='o',
            edgecolor='black', linewidth=2, alpha=0.95)

plt.colorbar()  # show color scale
plt.title("Simple Scatter with Color Mapping")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.show()
```

✅ What happens here:

* Point `(1, 4)` gets color mapped from value `10` → **lightest green**
* Point `(2, 5)` from value `5` → **medium green**
* Point `(3, 6)` from value `1` → **darkest green**



**6. Adding colormap with the size of the marker:**

```python
x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]

colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]

sizes = [209, 486, 381, 255, 191, 315, 185, 228, 174,
         538, 239, 394, 399, 153, 273, 293, 436, 501, 397, 539]

# 6. Adding marker edgecolor, edge width, transparency(alpha):
plt.scatter(x, y, s = sizes, c = colors, cmap='Greens', marker = "o",
            edgecolor = 'black', linewidth = 2, alpha = 0.95)

cbar = plt.colorbar() 
cbar.set_label("Satisfaction")
plt.title("Random values")
plt.xlabel("Values of x-axis")
plt.ylabel("Value of y-axis")
plt.show()
```

**7. Putting all together:**

```python

import pandas as pd
from matplotlib import pyplot as plt

plt.style.use('ggplot')

data = pd.read_csv('2019-05-31-data.csv')
view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']

plt.scatter(view_count, likes, c=ratio, cmap='Greens',
            edgecolor='black', linewidth=1, alpha=0.75)

cbar = plt.colorbar()
cbar.set_label('Like/Dislike Ratio')

# plt.xscale('log')
# plt.yscale('log')

plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.tight_layout()

plt.show()
```

```python

import pandas as pd
from matplotlib import pyplot as plt

plt.style.use('ggplot')

data = pd.read_csv('2019-05-31-data.csv')
view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']

plt.scatter(view_count, likes, c=ratio, cmap='Greens',
            edgecolor='black', linewidth=1, alpha=0.75)

cbar = plt.colorbar()
cbar.set_label('Like/Dislike Ratio')

plt.xscale('log')
plt.yscale('log')

plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.tight_layout()

plt.show()
```
