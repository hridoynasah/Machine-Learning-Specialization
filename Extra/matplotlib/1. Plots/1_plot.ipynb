**Here's a list of all the topics covered in the matplotlib demo:**

1. Plot the data to x and y axis: `plt.plot(x, y)`

2. Adding labels to the data: `plt.title()`, `plt.xlabel()`, `plt.ylabel()`

3. Adding legends to identify plot: `plt.legend()` or `plt.plot(..., label="name")`

4. Format Strings `fmt = '[marker][line][color]'` (e.g., `"k--"`, `"b-x"`)

5. Adding linewidth: `plt.plot(..., linewidth=3)`

6. Adding grid: `plt.grid(True)`

7. Check available plot styles: `plt.style.available`

8. Adding plot style: `plt.style.use("style_name")` (e.g., "fivethirtyeight", "ggplot")

9. Comic style plotting and Save as png:
- Comic style: `plt.xkcd()`
- Save plot: `plt.savefig("plot.png")`
